## Feature Extraction

```python
'''
 5 features:

Zero Crossing Rate
Chroma_stft
MFCC
RMS(root mean square) value
MelSpectogram to train our model.
'''
```

```python
import numpy as np
import librosa

def noise(data):
    noise_amp = 0.035 * np.random.uniform() * np.amax(data)
    return data + noise_amp * np.random.normal(size=data.shape[0])

def shift(data):
    shift_range = int(np.random.uniform(low=-5, high=5) * 1000)
    return np.roll(data, shift_range)

def pitch_augment(data, sampling_rate, pitch_factor=0.7):
    return  librosa.effects.pitch_shift(data, sr=sampling_rate, n_steps=pitch_factor, bins_per_octave=12, res_type='kaiser_best')


def extract_features(data, sample_rate):
    result = np.array([])

    # Zero Crossing Rate
    zcr = np.mean(librosa.feature.zero_crossing_rate(y=data).T, axis=0)
    result = np.hstack((result, zcr))

    # Chroma STFT
    stft = np.abs(librosa.stft(data))
    chroma_stft = np.mean(librosa.feature.chroma_stft(S=stft, sr=sample_rate).T, axis=0)
    result = np.hstack((result, chroma_stft))

    # MFCC
    mfcc = np.mean(librosa.feature.mfcc(y=data, sr=sample_rate, n_mfcc=13).T, axis=0)
    result = np.hstack((result, mfcc))

    # Root Mean Square
    rms = np.mean(librosa.feature.rms(y=data).T, axis=0)
    result = np.hstack((result, rms))

    # Mel Spectrogram
    mel = np.mean(librosa.feature.melspectrogram(y=data, sr=sample_rate).T, axis=0)
    result = np.hstack((result, mel))

    return result

def get_features(path):
    # Load original audio
    data, sample_rate = librosa.load(path, duration=2.5, offset=0.6)
    
    # Feature from original
    res1 = extract_features(data, sample_rate)
    result = np.array([res1])  # shape (1, features)

    # Feature from noise-augmented
    noise_data = noise(data)
    res2 = extract_features(noise_data, sample_rate)
    result = np.vstack((result, res2))

    # Feature from time-shifted
    shifted_data = shift(data)
    res3 = extract_features(shifted_data, sample_rate)
    result = np.vstack((result, res3))

    # Feature from pitch-shifted
    pitched_data = pitch_augment(data, sample_rate, pitch_factor=0.7)
    res4 = extract_features(pitched_data, sample_rate)
    result = np.vstack((result, res4))

    return result
```

```python
import pandas as pd
import numpy as np

# Load CSV
df = pd.read_csv("data_path.csv")  # Replace with actual path if needed

X, y = [], []

for i, row in df.iterrows():
    path = row['Path']
    label = row['Emotions']
    
    print(f"Iteration {i}: Path={path}, Label={label}")
    
    features = get_features(path)  # shape (4, N)
    
    for f in features:
        X.append(f)
        y.append(label)


X = np.array(X)
y = np.array(y)
```

## Model Setup and Training

```python
from tensorflow.keras.layers import Bidirectional, LSTM

model = Sequential()

model.add(Conv1D(256, kernel_size=5, strides=1, padding='same', activation='relu', input_shape=(X_train.shape[1], 1)))
model.add(BatchNormalization())
model.add(MaxPooling1D(pool_size=5, strides=2, padding='same'))

model.add(Conv1D(256, kernel_size=5, strides=1, padding='same', activation='relu'))
model.add(BatchNormalization())
model.add(MaxPooling1D(pool_size=5, strides=2, padding='same'))

model.add(Conv1D(128, kernel_size=5, strides=1, padding='same', activation='relu'))
model.add(BatchNormalization())
model.add(MaxPooling1D(pool_size=5, strides=2, padding='same'))
model.add(Dropout(0.2))

model.add(Conv1D(64, kernel_size=5, strides=1, padding='same', activation='relu'))
model.add(BatchNormalization())
model.add(MaxPooling1D(pool_size=5, strides=2, padding='same'))

# Adding Bidirectional LSTM layer here
model.add(Bidirectional(LSTM(64, return_sequences=False)))
model.add(Dense(units=32, activation='relu'))
model.add(Dropout(0.3))

model.add(Dense(units=8, activation='softmax'))
model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])

model.summary()
```
